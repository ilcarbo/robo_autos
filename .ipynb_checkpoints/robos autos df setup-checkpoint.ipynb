## Dataset import and setup

```python
# leer primer csv para ir agregando a este. Parsear como fechas las columnas 'tramite_fecha' y 'tramite_inscrip...'
robos_full = pd.read_csv('robos18/dnrpa-robos-recuperos-autos-201801.csv', parse_dates= ['tramite_fecha', 'fecha_inscripcion_inicial'])
```

```python
# importar y concatenar files 2018
path = './robos18'
all_files = glob.glob(path + "/*.csv")

df_list18 = []
for file in all_files:
    df = pd.read_csv(file, parse_dates= ['tramite_fecha', 'fecha_inscripcion_inicial'])
    df_list18.append(df)
    
robos_total18 = pd.concat(df_list18, axis = 0, ignore_index= True)
```

```python
np.shape(robos_total18)
```

```python
# importar y concatenar files 2019
path = './robos19'
all_files = glob.glob(path + "/*.csv")

df_list19 = []
for file in all_files:
    df = pd.read_csv(file, parse_dates= ['tramite_fecha', 'fecha_inscripcion_inicial'])
    df_list19.append(df)

# hay un df (el anteúltimo de 2019) que tiene diferentes los nombres de las dos últimas columnas
    
robos_total19 = pd.concat(df_list19, axis = 0, ignore_index= True)
```
